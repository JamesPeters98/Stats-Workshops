```python
from prettytable import PrettyTable
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t, chi2  # load normal pdf functions from scipy.stats


def PlotCI(prob=0.6827,mean=0,sigma=1,doubleTail=True):
    """
plotCI draws the exlcuded regions for a normal pdf (i.e. Gaussian)
prob = CI probability limit
mean = mean of normal PDF
sigma = sigma of normal PDF
doubleTail = is the distribution assumed to be two tailed
    """
    n=500 # number of points to draw, all pdf
    nfill=100 # number of points to draw in excluded region
    xmin=mean-5*sigma # set max and min of plot
    xmax=mean+5*sigma;
    x=np.linspace(xmin,xmax,n) # x of points
    y=norm.pdf(x,mean,sigma) # PDF of x values 
    # test if 1 or two tailed
    if doubleTail : # CI is the middle of the distribution 
        prob2=(1-prob)/2 # half above and half below prob
        # norm.isf is inverse survival function 
        # The survival function is defined as 1-cdf
        # i.e. the cumlative probaility that the real value > x
        # isf is used to convert a cumlative probility to x value
        xcut1  = norm.isf(1-prob2,mean,sigma) # lower limit in x within CI window
        xcut2  = norm.isf(prob2,mean,sigma) # upper limit in x within CI window
        xfill1 = np.linspace(x[0],xcut1,nfill) # points between xmin and xcut1 to plot as filled
        yfill1 = norm.pdf(xfill1,mean,sigma) # y values for xfill1
        # now need to add a point on the axis so the filled shape goes to zero
        xfill1 = np.append(xfill1,xcut1) # have to explicitly copy then extend a numpy array
        yfill1 = np.append(yfill1,0)
        xfill2 = np.linspace(xcut2,x[-1],nfill) # x[-1] as an index means last entry in x 
        yfill2 = norm.pdf(xfill2,mean,sigma)
        xfill2 = np.append(xcut2,xfill2) # here need to prepend the values for upper side band
        yfill2 = np.append(0,yfill2)
        plt.plot(x,y,label="pdf")
        plt.fill(xfill1,yfill1,'r',label="Lower excluded region CI={}".format(prob));
        plt.fill(xfill2,yfill2,'b',label="Upper excluded region CI={}".format(prob));
        plt.xlabel('x norm.pdf(x,$\mu={}, \sigma={}$)'.format(mean,sigma))
        plt.legend()
        plt.show()
    else:  # one tailed so can use CDF directly
        # use 1-prob as inverse survival function (sf = 1-cdf)
        xcut=norm.isf(1-prob,mean,sigma) # find upper side band limit in x
        xfill=np.linspace(xcut,x[-1],nfill) # make x points to USB limit
        yfill=norm.pdf(xfill,mean,sigma) # y values matching x values
        xfill = np.append(xcut,xfill) # have to explicitly copy then extend a numpy array
        yfill = np.append(0,yfill)
        plt.plot(x,y,label="pdf")
        plt.fill(xfill,yfill,'b',label="Upper excluded region CI={}".format(prob));
        plt.xlabel('x for norm.pdf(x,$\mu={}, \sigma={}$)'.format(mean,sigma))
        plt.legend()
        plt.show()
# You can use the following code to test the PlotCI function
#PlotCI(0.6827,0,1,True) # example to test code, double sided
#PlotCI(prob=0.95,mean=100,sigma=25,doubleTail=False) # example to test code, signle sided



figure = plt.figure(figsize=(6,4))

m = 10
s = 2.5

# CUMULATIVE DENSITY FUNCTION & GAUSSIAN PDF
x = np.linspace(m-5*s,m+5*s,100)
CDF = norm.cdf(x,m,s)
PDF = norm.pdf(x,m,s)
plt.plot(x,CDF, label="CDF")
plt.plot(x,PDF, label="PDF")
plt.xlabel('x')
plt.ylabel("probability")
plt.title("CDF & PDF for a Gaussian Distribution $\mu={}, \sigma={}$".format(m,s))
plt.legend()
plt.show()
```

The next plots show the two-tailed Confidence Intervals for 68.27% and 95% 


```python

# Return upper and lower Confidence Interval
def calculateCI(CI,mean,sigma,doubleTail=True):
    if doubleTail: prob = (1-CI)/2
    else: prob = 1-CI
    return norm.isf(1-prob,mean,sigma),norm.isf(prob,mean,sigma)

# Display Confidence Interval & return the upper and lower values.
def displayCI(CI,mean,sigma,doubleTail=True):
    l,u = calculateCI(CI,mean,sigma,doubleTail)
    PlotCI(CI,mean,sigma,doubleTail)
    plt.show()
    return CI,l,u


r = displayCI(0.6827,m,s)
r2 = displayCI(0.95,m,s)

table = PrettyTable(['Confidence Interval','Lower Bound','Upper Bound'])
table.add_row(["{0:1.2f}%".format(r[0]*100),"{0:1.2f}".format(r[1]),"{0:1.2f}".format(r[2])])
table.add_row(["{0:1.2f}%".format(r2[0]*100),"{0:1.2f}".format(r2[1]),"{0:1.2f}".format(r2[2])])
print(table)
```

The next plot shows the one-tailed 95% upper CI. This is the value where 95% of the distribution lies below, from $-\infty$ to $x$

```python
r4 = displayCI(0.95,m,s,False)
print("The Upper CI = {0:1.2f}".format(r4[2]))
```

# Question 2

The first plot shows the 95% Confidence Interval for $\mu = 16$

The next two plots show the 95% & 99% Confidence Intervals for $\mu = 16.6$

```python
## Returns the t-value for a given sample mean
def testStatistic(sample_mean,mean,standard_deviation,n):
    return (sample_mean-mean)/(standard_deviation/np.sqrt(n))

## Returns the probability of a sample mean being part of the given distribution.
def probability(sample_mean,mean,standard_deviation,n,doubleTail=True):
    dt = 1
    if doubleTail: dt = 2
    t_value = testStatistic(sample_mean,mean,standard_deviation,n)
    return t.sf(t_value,n-1)*dt

## Boolean value if the given probability lies within the given significance level.
def rejectNullHypothesis(confidence_interval,probability):
    p = (1-confidence_interval)
    if probability < p:
        return False
    return True

## VARIABLES
n = 100     # Number of trials
m = 16      # Population mean
s = 3       # Standard Deviation

CI = 0.95   # Confidence Interval
CI2 = 0.99   # Confidence Interval
m_sample = 16.6   # Sample mean

## Display graphs
result = displayCI(CI,m,s)
displayCI(CI,m_sample,s)
displayCI(CI2,m_sample,s)

# Calculate probablity for the given sample mean, and whether the hypothesis
# is true for given confidence intervals.
p = probability(m_sample,m,s,n,True)
valid_95 = rejectNullHypothesis(CI,p)
valid_99 = rejectNullHypothesis(CI2,p)

print("Table showing Null Hypothesis' for a sample mean {0} with population mean {1}".format(m_sample,m))
table = PrettyTable(['Confidence Interval','Probability','Reject Hypothesis'])
table.add_row(["{0:1.2f}%".format(CI),"{0:1.4f}".format(p),"{0}".format(valid_95)])
table.add_row(["{0:1.2f}%".format(CI2),"{0:1.4f}".format(p),"{0}".format(valid_99)])
print(table)
```

# Question 3

```python
from scipy.stats import chi2

## VARIABLES
n = 18
dof = n-1
chi = 29.6

# Reduced Chi^2 Calculation.
rChi2 = chi/dof
print("Reduced Chi^2 = {0:1.4f}".format(rChi2))

# Probability of this Chi^2 for the given D.o.F.
p = chi2.sf(chi,dof)

# Null Hypothesis' for 95% & 99%
valid_95 = rejectNullHypothesis(0.95,p)
valid_99 = rejectNullHypothesis(0.99,p)


# Table of results.
print("")
print("Table showing Null Hypothesis' for chi^2 = {0}".format(chi))
table = PrettyTable(['Confidence Interval','Probability','Reject Hypothesis'])
table.add_row(["{0:1.2f}%".format(0.95),"{0:1.4f}".format(p),"{0}".format(valid_95)])
table.add_row(["{0:1.2f}%".format(0.99),"{0:1.4f}".format(p),"{0}".format(valid_99)])
print(table)

def plotChiDist(chi,dof,CI):
    # Plot graphs.
    x = np.linspace(0,2.5*dof,100)
    
    # Calc upper lower limits
    upper_limit = chi2.isf((1-CI),dof)
    
    # Pdf calcs
    CDF = chi2.pdf(x,dof)           # Curve
    u = chi2.pdf(upper_limit,dof)   # Upper probability
    y_chi2 = chi2.pdf(chi,dof)      # Sample Chi^2
    
    # Plot curve and points
    plt.plot(x,CDF, label="$\chi^2$ Distribution")
    plt.plot(upper_limit,u,'k*', label="Critical Limit")
    plt.plot(chi,y_chi2,'r*', label="$\chi^2$" )
    plt.title("Plot Showing the $\chi^2$ Distribution with a {}% Confidence Interval".format(CI*100))
    
    plt.xlabel('$\chi^2$')
    plt.ylabel("probability")
    plt.legend()
    plt.show()
    
plotChiDist(chi,dof,0.95)
plotChiDist(chi,dof,0.99)



```
